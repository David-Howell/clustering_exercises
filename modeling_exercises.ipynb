### - c. Use the elbow method to visually select a good value for k.

```python
kmeans.inertia_
```

```python
{f'k = {x}': x for x in range(6)}
```

```python
pd.Series(
    {k: KMeans(k).fit(X_train).inertia_
    for k in range(1,10)}).plot(marker='x')
plt.ylabel('Inertia')
plt.xlabel('k number of clusters')
plt.title('Change in Inertia as K increases')
plt.show()
```

```python
inertia_series = pd.Series(
    {k: KMeans(k).fit(X_train).inertia_
    for k in range(1,10)})
```

```python
inertia_series
```

```python
for i in range(1, len(inertia_series)):
    x = (inertia_series[i] - inertia_series[i+1])
    print(f'For n_clusters = {i}: Inertia is = {round(x,2)}')
    if x < 1:
        print(f'The elbow is {i}')
        kmeans = KMeans(i).fit(X_train)
        print(kmeans)
        train['cluster'] = kmeans.predict(X_train)
        break
        
```

```python
for i, j in enumerate(X_cols):
    for k in range(i+1, len(X_cols)):
        sns.relplot(data=train,
                    x= j,
                    y= X_cols[k],
                    hue= 'species_name',
                    col= 'cluster'
                   )#.set(
#             title= f'{j.capitalize()} vs. {X_cols[k].capitalize()}, colors = Species')
```

### - d. Repeat the clustering, this time with 3 different features.

```python
for i, j in enumerate(X_cols):
    for k in range(i+1, len(X_cols)):
        
        double_var = [j, X_cols[k]]

        inertia_series = pd.Series(
            {k: KMeans(k).fit(train[double_var]).inertia_
            for k in range(1,10)})
        
        inertia_series.plot(marker='x')
        plt.ylabel('Inertia')
        plt.xlabel('k number of clusters')
        plt.title('Change in Inertia as K increases')
        plt.show()


        for l in range(1, len(inertia_series)):
            y = (inertia_series[l])
            x = (inertia_series[l] - inertia_series[l+1])
            print(f'For n_clusters = {l}: Inertia is = {round(y,2):>4}: diff = {round(x,2)}')
            elbow = input('Is this the elbow? [n]/y ')
            if elbow == 'y':
                print(f'The elbow is {l}')
                kmeans = KMeans(l).fit(train[double_var])
                print(kmeans)
                train[f'cluster_{j}_{X_cols[k]}'] = kmeans.predict(train[double_var])
                print(train[double_var].columns.to_list())
                print(f'cluster_{j}_{X_cols[k]}')
                
#                 sns.relplot(data=train,
#                             x= j,
#                             k= X_cols[k],
#                             hue= 'species_name',
#                             col= f'cluster_{j}_{X_cols[k]}'
#                            )
                
                break
        
```
